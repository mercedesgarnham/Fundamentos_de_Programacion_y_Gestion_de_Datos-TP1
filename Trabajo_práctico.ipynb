```python
#1. Lectura del Dataset

import pandas as pd

url = "https://raw.githubusercontent.com/mercedesgarnham/Fundamentos_de_Programacion_y_Gestion_de_Datos-TP1/main/IRIS.csv"
df_iris = pd.read_csv(url)
```

```python
#2. Descripción General del Dataset
print("################################")
print("# Primeras 5 filas del dataset #")
print("################################")
print("\n")

print(df_iris.head(),"\n")
print("\n")

print("###################################")
print("# Información general del dataset #")
print("###################################")
print("\n")

print(df_iris.info(),"\n")
print("\n")

print("#############################")
print("# Estadísticas descriptivas #")
print("#############################")
print("\n")

print(df_iris.describe(),"\n")
print("\n")
print("#############################################")
print("# Cantidad de valores faltantes por columna #")
print("#############################################")
print("\n")
print(df_iris.isnull().sum(),"\n")
print("\n")
```

```python
#3. Análisis por Especie: Conteo y Promedio de Características
print("**************************************************************")
print("Cantidad de registros para cada especie presente en el dataset")
print("\n")
for especie in df_iris['species'].unique():
  contador = 0
  for valor in df_iris['species']:
    if valor == especie:
      contador += 1
  print(f"El número de registros para la {especie} es {contador}.")
print("\n")

# Creamos la función calcular promedio
def calcular_promedio(lista_numeros):
    suma = 0
    for numero in lista_numeros:
        suma += numero
    if len(lista_numeros) == 0:
      promedio = 0
    else:
      promedio = suma / len(lista_numeros)
    return promedio

print("****************************************************************")
print("Calculo de promedio por especie de las características numéricas")
print("\n")

for especie in df_iris['species'].unique():
  print("Datos de la especie",especie)
  print("--------------------------------------")
  datos_especie = df_iris[df_iris['species'] == especie]
  columnas_numericas = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
  for columna in columnas_numericas:
    promedio = calcular_promedio(datos_especie[columna])
    print(columna," = ",promedio)
  print("\n")
```

```python
#4. Almacenamiento de Promedios por Especie
promedios_especies = [] # Creamos una lista donde se va a guardar un diccionario por especie
columnas_numericas = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for especie in df_iris['species'].unique():
  datos_especie = df_iris[df_iris['species'] == especie]
  promedios_especie = {'especie': especie} # inicializamos el diccionario

  for columna in columnas_numericas:
    promedio = calcular_promedio(datos_especie[columna]) # calculamos el promedio para cada característica
    promedios_especie[columna] = promedio # guardamos el promedio en el diccionario

  promedios_especies.append(promedios_especie) # agregamos el diccionario con promedios a la lista

print(promedios_especies)
```

```python
#5. Comparación Interactiva de Características entre Especies


def comparar_caracteristica():
  # Imprimimos las opciones
  print("Opciones de características: ",columnas_numericas)
  # Pedimos al usuario la característica a comparar
  opcion = input("\n¿Qué característica deseas comparar? (Ingresa el nombre): ").strip().lower()
  print("\n")
  if opcion == "sepal_length":
    caracteristica = "sepal_length"
  elif opcion == "sepal_width":
    caracteristica = "sepal_width"
  elif opcion == "petal_length":
    caracteristica = "petal_length"
  elif opcion == "petal_width":
    caracteristica = "petal_width"
  else:
    print("Opción no válida")

  valor_maximo = 0

  # Recorremos los promedios almacenados en el punto 4.
  for datos_especie in promedios_especies:
    especie = datos_especie['especie']
    promedio = datos_especie[caracteristica]
    if promedio > valor_maximo:
      valor_maximo = promedio
      especie_max = especie

  return f"La especie {especie_max} tiene el valor máximo promedio de {valor_maximo} para la característica '{caracteristica}'"

print(comparar_caracteristica())
```

```python
#6. Búsqueda y Conteo de un Dato Específico

def buscar_y_contar(df):
    # pedimos al usuario que ingrese el valor a buscar
    valor_buscar = input(f"¿Qué valor querés buscar en la columna? ")
    print("valor a buscar",valor_buscar)

    # recuperamos las opciones de columnas
    columnas = df.columns.tolist()
    print("Opciones de columnas: ",columnas)
    # pedimos al usuario que ingrese la columna a buscar
    columna = input("\n¿En qué columna querés buscar ese valor? (Ingresá el nombre exacto): ").strip()
    print("Columna a buscar",columna)

    if columna in df.columns:
      #valor_buscar = float(valor_buscar)
      contador = 0
      # Recorremos la columna especificada
      for valor in df[columna]:
        if str(valor) == valor_buscar:
          contador += 1
    else:
      print(f"La columna '{columna}' no existe.")

    print("\n")
    return f"El valor: '{valor_buscar}' se encuentra {contador} veces en la columna '{columna}'."

# Ejecutar la función
buscar_y_contar(df_iris)
```

```python
#7. Cálculo de Suma Total

def calcular_suma_total(df):
    # Mostramos las columnas numéricas disponibles
    print("Columnas numéricas disponibles:", columnas_numericas)
    # Pedimos al usuario que ingrese el nombre de la columna numérica
    columna = input("\n¿Qué columna numérica querés sumar? (Ingresá el nombre exacto): ")
    print("Columna seleccionada:", columna)

    if columna in df.columns:
        # inicializamos suma_total y recorremos la columna usando un for
        suma_total = 0
        for valor in df[columna]:
            suma_total += valor
    else:
        print(f"La columna no existe.")
        suma_total = 0
    print("\n")
    return f"La suma total de la columna '{columna}' es: {suma_total}"

# Ejecutar la función
calcular_suma_total(df_iris)
```
